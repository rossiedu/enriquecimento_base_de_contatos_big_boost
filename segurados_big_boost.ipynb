# Criando a Classe BuscaTelefone

```python
class BuscaTelefone:

  def __init__(self,cpf):
    cpf = str(cpf)
    cpf = cpf.replace(".",'')
    cpf = cpf.replace('-','')
    cpf = cpf.zfill(11)
    self.cpf = cpf

  def BuscaPhone(self):
    url = "https://bigboost.bigdatacorp.com.br/peoplev2"
    accessToken = 'token'
    payload = json.dumps({"Datasets": "phones_extended","q": "doc"+"{"+f"{self.cpf}"+"}","AccessToken": f"{accessToken}"}) 
    headers = {'Content-Type': 'application/json'}
    response = requests.request("POST", urlPeople, headers=headers, data=payload)
    d = str(response.json().get('Result'))
    #Fatiamento de String para Achar o Numero de Telefone
    i = d.find('Number')+10 # inicio
    f = d.find('AreaCode')-4 # fim
    Phone = d[i:f]
    #Fatiamento de String para Achar o DDD
    s = d.find('AreaCode')+12 # start
    e = d.find('CountryCode')-4 # end
    DDD = d[s:e]
    #Fatiamento de String para Achar o CountryCode
    c = d.find('CountryCode')+15 # comeca
    t = d.find('Complement')-4 # termina
    CountrCode = d[c:t]
    CompletedPhone = CountryCode + DDD + Phone 
    return CompletedPhone

  def __repr__(self):
    return self.BuscaPhone()

  

```
